```python
# function to calculate curve confidence intervals

def confidence_interval(a,b,c,d,e):
    multiple_lists = [a,b,c,d,e]
    arrays = [np.array(x) for x in multiple_lists]
    mean_curve = [x* 100 for x in [np.mean(k) for k in zip(*arrays)]]
    # max_curve = [np.max(k) for k in zip(*arrays)]
    # min_curve = [np.min(k) for k in zip(*arrays)]
    max_curve = [x* 100 for x in[np.mean(k) + np.std(k) for k in zip(*arrays)]] #use std. dev instead of max
    min_curve = [x* 100 for x in[np.mean(k) - np.std(k) for k in zip(*arrays)]] #use std. dev instead of min

    #index of max of list
    max_index = mean_curve.index(max(mean_curve))
    max_std = [np.std(k) for k in zip(*arrays)][max_index] #use std. dev instead of min
    return mean_curve, max_curve, min_curve, max_std
```
